#Standardization 
*   For each feature/column, scale the values so the mean is 0 and the standard deviation is 1.

#Normalization 
*   Rescale the range of each feature to be a different range, usually [0, 1]
*   The smallest value would be 0 and the largest value would be 1. 
Screen Shot 2021-04-10 at 3.44.57 PM.png



#Robust Scaling 
*   Robust scaling: the same as normalization, but outliers are not considered when scaling teh features
*   Scaled x = (x - median(x)) / interquartile range where x is a feature vector 
*   Robust scaling does not remove outeliers from the data, but does not allow them to skew the scaling of the other, non-outlier values
*   This prevents the outliers from hindering the performance of machine learning algorithms







1.   Standardization assumes the original features have Gaussian distributions, just on different scales
2.   If original data has Gaussian or Gaussian-like distribution, then standardization nicely preserves the original structure of your data
3.   If features have Gaussian-like distributions and algorithm assumes a Gaussian distribution, standardization = better performance 
4.   Normalization is better for when you don't know the feature distributions or you know they aren't Gaussian
5.   Most modern algorithms (i.e. neural networks) do not assume a Gaussian distribution, so normalization is a safe bet!

#Principle Component Analysis (PCA)
* Goal of PCA is to provide a few features that are combinations of the original features
* These features are some linear combination of each other
* These features should have maximized variance (basically ow we can best spread out the data)
* Data which is more spread out has more information which could be valuable to us in our projects

#Covariance Matrix

* Covariance matrices contain covariances between the pairs of features. 
* Covariance matrices compare the variability of two features.

* Matrices have eigenvectors which are vectors whose directions do not change when linear transformations are applied to them. 
* These vectors can be stretched out but still accurately represent the data 
* Largest eigenvector of the covariance matrix (linear transformations of the original data) is the most significant principle component 
* We make this one of the axes of our new, transformed dataset
* Second largest eigenvector is second most significant principle component

#UMAP 
* UMAP - like t-SNE but faster and somewhat preserves global strucutre too 
* t-SNE does not preserve global structure
* UMAP tries to figure out neighborhoods of data points (data points that are close together)
* UMAP then uses a special distance function to recreate the same structure in a small amount of dimensions (typically 2-3)
* Summary: UMAP makes a graph of the data then optimizes a low-dimensional graph to be as similar as possible 

1. First, find the probabilities that points are connected (greater distance away = lower probabilities) 
2. Next step is to optimize low-dimensional graph (the hard part)
3. 2 key parameters: n_neighbours and min_dist
4. n_neighbours is the number of nearest neigfhbors used to make the original hi-dimnesional graph 
5. min_dist is the minimum allowed distance between points in the low-dimensional space 
6. Some excellent interactive visualizations of UM AP: https://pair-code.github.io/understanding-umap/

* UMAP uses a loss function to minimize the error when comparing connections (simplices) between points in the high-dimensional and low-dimnsional space

* UMAP compares the probabilities of two simplices existing and optimizes the structure of the low-dimensional data based on the divergence 

* N_neighbours controls the interconnectedness of the high-dimensional data which directly controls how global the optimized low-D simplices will be 

* This is customized by the user and should depend on the data 

* Min_dist controls how tightly UMAP can pack points to reduce the loss- this is also customized by the user and should depend on the data

1. PCA - used when tasks lack nuance and trying to find somehting linear 

2. UMAP is more flexible, but important to adjust parameters correctly 

3. Visualizing cell differentation, global structure is excellent for identifying rarae/transitional populations and n_neighbours should be higher
