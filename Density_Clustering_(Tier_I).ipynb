#Density CLustering (Tier I)
1. Method of finding clusters is based on density 
2. Find regions of data that are very dense - those regions are the clusters 
3. Once we have the dense 'epicenter' of the data, we can assign other points to the center closest to it
4. Entire datasets get labeled.

#DBSCAN 
1. You can actually check every point 
2. Density-based spatial clustering of applications with noise (DBSCAN) does this 
3. CLuster 'centers' are points within a certain number of points within a given radius of it 
4. Points in that cluster are points where tehre is a path to that point containing only centers 
4. All others are outliers 

#Kernel Density Estimation
1. KDE fits some kernel (i.e. Gaussian) at each data point
2. To estimate the probability that a point is at a location, see how many points nearby, assigns that point a probability value 
3. Used in clustering to estimate the highest density regions as cluster centers and assign other points from there 
4. https://mathisonian.github.io/kde/
   

#Nearest Neighbor Searching
1. How many points are near a point 
2. Approximate nearest neighbor approaches reduce runtime while preserving accuracy by guessing where neighborhoods are 
3. Locality sensitive hashing is an important (but straightforward) technique here
4. Basic idea: put data that close together in the same buckets and check to see how many points are in which buckets
5. Flexible with many different metrics for 'closeness'
