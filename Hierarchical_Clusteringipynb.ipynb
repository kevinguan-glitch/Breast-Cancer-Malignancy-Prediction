#Hierarchical Clustering (TIER II)
* We need effective methods to group data into clusters, even if data is not labeled. The burden for interpretation is on the programmer. 

#Hierarchical Clustering
1. One method for finding clusters is based on hierarchies within the data 
2. Two primary subtypes of hierarchical clustering: agglomerative and divisive 

#Agglomerative Clustering
* Make every data point its own cluster (bottom-up) 
* Using some metric, check to see which clusters are closest together 
* Merge those ones 
* Keep doing that until some threshold/standard is met

#Agglomerative clustering works through linkage 
#Linkage is how the algorithms determine which clusters are closest together 
#The type of linkage can determine how fast or how accurate your clustering performance is

#Linkages
* Single linkage: use the smallest possible distance between the 2 clusters 
  * Heavily influenced by outliers + clusters that are not nicely circular 

* Maximum Linkage: Use the largest possible distance between the 2 clusters 
  * Heavily influenced by outliers + clusters that are not nicely circular 

* Ward: Minimize the increase in variance when merging two clusters (preserve uniformity)

* Average: Average distance between two sets 

#Computing Linkage Distance
* Can use Euclidean, Manhattan, Chebyshev 
* Can use other distance metrics like Levenshtein distance depending on the application

#Divisive Clustering
* Opposite of agglomerative clustering: starts with one big cluster
* Split the cluster into multiple smaller clusters using a secondary clustering algorithm (i.e. K-Means)
* Take the most dissimilar of the smaller clusters and split that one using the secondary algorithm 
* Do this until some threshold/standard is met 

1. Use same linkages and linkage distance metrics to determine which clusters need to be split 
2. The difference is that this happens within a cluster instead of between clusters
3. What is the largest possible distance you can have within a cluster? What is the cluster with the most internal variance? 
4. Once you identify the most dissimilar cluster, use a secondary "flat" (non-hierarchical) clustering algorithm to break it down
5. Continue this until some threshold is met (i.e. 10 cluster example)

#Divisive Versus Agglomerative 
1. Divisive clustering requires a separate "flat" clustering algorithm to split up the big clusters (i.e. K-means)

2. This makes it more complex/can be slower (unless you do not need to go very far down the hierarchy) 

3. HOwever, considering the global structure when making decisions makes the clusters more accurate 

4. Agglomerative clustering can merge clusters which appear close together but belong in different populations relative to the global structure

5. Agglomerative clustering is advanttageous for situtaiotns when a second clustering algoirthm is not feasible (linkages are simple)
